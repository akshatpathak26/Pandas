```python
points_df = pd.DataFrame(
    {
        '1st point' : [(1,3), (3,4),(4,-5),(-5,10),(10,12)],
        '2nd point' : [(-1,3), (-3,4),(4,-5),(-5,10),(10,-12)]
    }
)
```

```python
points_df
```

```python
def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    
    return ((pt_A[0] - pt_B[1]) ** 2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance']=points_df.apply(euclidean,axis = 1)
```
